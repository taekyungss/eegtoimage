```python
import numpy as np
import os
from glob import glob

train_folder = 'eeg_imagenet40_cvpr_2017_raw/train'

npy_files = glob(os.path.join(train_folder, '*.npy'))

# 음수 값 확인
for file in npy_files:
    data = np.load(file, allow_pickle=True)
    if data[1].min() < 0:
        print(f"음수 ㅇ: {file}")
        print(data[1][data[1] < 0])
    else:
        print(f"음수 x: {file}")
```
